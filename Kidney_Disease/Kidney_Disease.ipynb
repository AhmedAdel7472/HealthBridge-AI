#### EDA

```python
def voilin(col):
    fig  = px.violin(df, y=col, x='class', color='class', box=True, template='plotly_dark')
    return fig.show()

def kde(col):
    grid = sns.FacetGrid(df, hue='class', height = 6, aspect = 2)
    grid.map(sns.kdeplot, col)
    grid.add_legend()
    
def scatter_plot(col1, col2):
    fig  = px.scatter(df, x=col1, y=col2, color="class",  template='plotly_dark')
    return fig.show()

kde('red_blood_cell_count')
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df[cat_cols].isnull().sum()
```

```python
df.head(10)
```

```python
def random_sampling(feature):
    random_sample = df[feature].dropna().sample(df[feature].isna().sum())
    random_sample.index = df[df[feature].isnull()].index
    df.loc[df[feature].isnull(), feature] = random_sample

def impute_mode(feature):
    mode = df[feature].mode()[0]
    df[feature] = df[feature].fillna(mode)
```

```python
# random sampling for numerical value
for col in num_cols:
    random_sampling(col)
```

```python
df[num_cols].isnull().sum()
```

```python
random_sampling('red_blood_cells')
random_sampling('pus_cell')

for col in cat_cols:
    impute_mode(col)
```

```python
df[cat_cols].isnull().sum()
```

#### Feature Encoding

```python
for col in cat_cols:
    print(f"{col} has {df[col].nunique()}")
```

```python
# label_encoder
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
X = df.drop('class', axis = 1)
y = df['class']
```

#### Decision Tree Classifier

```python
from sklearn.model_selection import train_test_split

X_train,X_test, y_train, y_test =  train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Initialize the DecisionTreeClassifier
dtc = DecisionTreeClassifier()

# Train the model
dtc.fit(X_train, y_train)

# Calculate training and testing accuracy
dtc_train_acc = accuracy_score(y_train, dtc.predict(X_train))
dtc_test_acc = accuracy_score(y_test, dtc.predict(X_test))

# Print results
print(f"Training Accuracy of DTC is {dtc_train_acc}")
print(f"Testing Accuracy of DTC is {dtc_test_acc}")

# Confusion Matrix
print(f"Confusion Matrix of DTC is \n{confusion_matrix(y_test, dtc.predict(X_test))}\n")

# Classification Report
print(f"Classification Report of DTC is \n{classification_report(y_test, dtc.predict(X_test))}")
```

```python
import pickle

# Save the model to a file
with open("decision_tree_model.pkl", "wb") as f:
    pickle.dump(dtc, f)

print("Model saved successfully as 'decision_tree_model.pkl'!")
```
